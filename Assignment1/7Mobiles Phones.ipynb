```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.amazon.in/Smartphones-%E2%82%B910-000-%E2%82%B920-Basic-Mobiles/s?rh=n%3A1805560031%2Cp_36%3A1318506031')
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
soup.title.text
```

```python
Product_Name=[] 
Price=[]
Image_URL=[] 
Average_Rating=[]
```

```python
product=soup.find_all('a',{'class':'a-link-normal a-text-normal'})
```

```python
product[1:10]
```

```python
for i in product:
    
    Product_Name.append(i.get_text().replace('\n',''))
Product_Name    
```

```python
product[0].get_text().replace('\n','')
```

```python
price=soup.find_all('span',{'class':'a-price-whole'})
```

```python
price[1:10]
```

```python
for i in price:
    
    Price.append(i.get_text().replace('\n',''))
Price  
```

```python
average=soup.find_all('div',{'class':'a-row a-size-small'})
```

```python
average[1:10]
```

```python
for i in average:
    
    Average_Rating.append(i.get_text().replace('\n',''))
Average_Rating
```

```python
print(len(Product_Name),len(Price),len(Average_Rating))
```

```python
import pandas as pd
Mobiles=pd.DataFrame({})
Mobiles['product']=Product_Name
Mobiles['price']=Price
Mobiles['average']=Average_Rating
```
