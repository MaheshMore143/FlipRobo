```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.monster.com/jobs/search/?q=Software-Developer&where=Australia&cy=AU')
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
soup.title.text
```

```python
job_titles=[]
company_names=[]
location_list=[]
days_posted=[]
```

```python
titles=soup.find_all('h2',{'class':'title'})
```

```python
titles[0:10]
```

```python
for i in titles:
    
    job_titles.append(i.get_text().replace('\n',''))
job_titles   
```

```python
titles[0].get_text().replace('\n','')
```

```python
companies=soup.find_all('div',{'class':'company'})
```

```python
companies[0:10]
```

```python
for i in companies:
    
    company_names.append(i.get_text().replace('\n',''))
company_names   
```

```python
location=soup.find_all('ul',{'class':'list-inline'})
```

```python
location[0:10]
```

```python
for i in location:
    
    location_list.append(i.get_text().replace('\n',''))
location_list   
```

```python
day=soup.find_all('div',{'class':'meta flex-col'})
```

```python
day[0:10]
```

```python
for i in day:
    
    days_posted.append(i.get_text().replace('\n',''))
days_posted   
```

```python
print(len(job_titles),len(company_names),len(location_list),len(days_posted))
```

```python
import pandas as pd
jobs=pd.DataFrame({})
jobs['titles']=job_titles
jobs['company']=company_names
jobs['location']=location_list
jobs['days']=days_posted
```
