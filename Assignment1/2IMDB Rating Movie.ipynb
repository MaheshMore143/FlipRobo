```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.imdb.com/list/ls091520106/?sort=user_rating,desc&st_dt=&mode=detail&page=1')
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
soup.title.text
```

```python
Movie_name=[]
Year_Ofrelease=[]
Rating=[]
```

```python
movie=soup.find_all('h3',{'class':'lister-item-header'})
```

```python
movie[0:100]
```

```python
for i in movie:
    
    Movie_name.append(i.get_text().replace('\n',''))
Movie_name
```

```python
year=soup.find_all('span',{'class':'lister-item-year text-muted unbold'})
```

```python
year[0:100]
```

```python
for i in year:
    
    Year_Ofrelease.append(i.get_text().replace('\n',''))
Year_Ofrelease
```

```python
rating=soup.find_all('div',{'class':'ipl-rating-star small'})
```

```python
rating[0:100]
```

```python
for i in rating:
    
    Rating.append(i.get_text().replace('\n',''))
Rating
```

```python
print(len(Movie_name),len(Year_Ofrelease),len(Rating))
```

```python
import pandas as pd
Movie=pd.DataFrame({})
Movie['name']=Movie_name
Movie['year']=Year_Ofrelease
Movie['rating']=Rating
```
