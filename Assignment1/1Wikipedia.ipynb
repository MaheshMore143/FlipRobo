```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
soup.title.text
```

```python
Header_tag=[]
```

```python
tag=soup.find_all('h2',{'class':'mp-h2'})
```

```python
tag[0:20]
```

```python
for i in tag:
    
    Header_tag.append(i.get_text().replace('\n',''))
Header_tag    
```

```python
tag[0].get_text().replace('\n','')
```

```python
print(len(Header_tag))
```

```python
import pandas as pd
tags=pd.DataFrame({})
tags['tag']=Header_tag
```
