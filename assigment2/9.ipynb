```python
import selenium
import pandas as pd
from selenium import webdriver
```

```python
driver=webdriver.Chrome(r'C:\Users\DELL\Downloads\chromedriver.exe')
```

```python
url='https://www.myntra.com/shoes?f=Color%3ABlack_36454f&plaEnabled=false&rf=Price%3A5899.0_11599.0_5899.0%20TO%2011599.0'
```

```python
driver.get(url)
```

```python
product_brand=[]
product_description=[]
product_price=[]
```

```python
brand_tag=driver.find_elements_by_xpath("//h3[@class='product-brand']")
```

```python
brand_tag[0:100]
```

```python
for i in brand_tag:
    brand=i.text
    product_brand.append(brand)
product_brand[0:100]    
```

```python
des_tag=driver.find_elements_by_xpath("//h4[@class='product-product']")
```

```python
des_tag[0:100]
```

```python
for i in des_tag:
    des=i.text
    product_description.append(des)
product_description[0:100]    
```

```python
price_tag=driver.find_elements_by_xpath("//div[@class='product-price']")
```

```python
price_tag[0:100]
```

```python
for i in price_tag:
    price=i.text
    product_price.append(price)
product_price[0:100]    
```

```python
print(len(product_brand),len(product_description),len(product_price))
```

```python
import pandas as pd
product=pd.DataFrame({})
product['Brand']=product_brand
product['Description']=product_description
product['Price']=product_price
```
