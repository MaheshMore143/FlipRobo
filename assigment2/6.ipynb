```python
import selenium
import pandas as pd
from selenium import webdriver
```

```python
import time
```

```python
driver=webdriver.Chrome(r'C:\Users\DELL\Downloads\chromedriver.exe')
```

```python
url='https://www.flipkart.com/search?q=sunglasses&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off'
```

```python
driver.get(url)
```

```python
product_brand=[]
product_price=[]
product_discount=[]
```

```python
for i in range(0,3):
    brand_tag=driver.find_elements_by_xpath("//div[@class='_2WkVRV']")
    for i in brand_tag:
        product_brand.append(i.text)
   
    price_tag=driver.find_elements_by_xpath("//div[@class='_30jeq3']")    
    for j in price_tag:
        product_price.append(j.text)
        
    discount_tag=driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']/span[1]")    
    for m in discount_tag:
        product_discount.append(m.text)
    driver.find_element_by_xpath("//a[@class='_1LKTO3']/span[1]").click()    
    time.sleep(3)
```
